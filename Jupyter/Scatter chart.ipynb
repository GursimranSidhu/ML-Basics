```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
userage=np.array([20,30,34,50,70])
salaries=np.array([10,30,20,40,50])
```

```python
plt.plot(userage,salaries)
```

```python
plt.plot(userage,salaries,marker=".",ms=12,mec="red",mfc="yellow")
```

```python
plt.plot(userage,salaries,marker=".",ms=12,mec="red",mfc="yellow",color="green",linewidth=2,linestyle="dashdot")
```

```python

plt.plot(userage,salaries,
         marker=".",
         ms=12,
         mec="red",
         mfc="yellow",
         color="green",
         linewidth=2,
         linestyle="dashdot")
plt.xlabel("Employee age")
plt.ylabel("Employee salaries")
plt.title("Employee Info",loc="left",color="blue")
plt.grid(linestyle="dashed")

# plt.annotate(xy=(40,40),text="hello")

for i in range(len(userage)):
    plt.annotate(xy=(userage[i],salaries[i]),text=f"{userage[i],salaries[i]}")
```

```python
month=np.array([2,4,5,7,4,6,7,3,8,6])
sales=np.array([200,400,300,500,600,700,200,500,700,900])
profit=[200,500,300,400,300,600,300,200,700,100]
# coor=["red","green","blue","yellow","black","gray","brown","pink","orange","lightgreen "]
coor = ["red", "green", "yellow", "pink", "blue", "orange", "gray", "lightgreen", "brown", "black"]
```

```python
plt.scatter(month,sales,sizes=profit,color=coor )
for i in range(len(month)):
    plt.annotate(xy=(month[i],sales[i]),text=f"{month[i],sales[i]}")
plt.grid(linestyle="dashed")
```

```python
# larger the dot larger is the profit
```
